```python
import pandas as pd
import numpy as np
from os.path import basename

def split_into_csv(excel_path: str):
    try:
        name = basename(excel_path)
        name = name.replace(".xlsx", "")

        df = pd.read_excel(excel_path, header=None)

        rows = list(range(3,30))
        
        s = df.loc[1, :]
        hs = s.where(s == "L").dropna().index.tolist()
        h0 = list(range(hs[0], hs[0]+4))
        h240 = list(range(hs[1], hs[1]+9))
        h530 = list(range(hs[2], hs[2]+9))

        df_h0 = pd.DataFrame(df.loc[rows,h0].values, index=df.loc[rows,0].fillna("Papel"), columns=df.loc[1,h0]).reset_index().rename(columns={0:'color'})
        df_h240 = pd.DataFrame(df.loc[rows,h240].values, index=df.loc[rows,0].fillna("Papel"), columns=df.loc[1,h240]).reset_index().rename(columns={0:'color'})
        df_h530 = pd.DataFrame(df.loc[rows,h530].values, index=df.loc[rows,0].fillna("Papel"), columns=df.loc[1,h530]).reset_index().rename(columns={0:'color'})

        
        df_h0["dL"] = np.nan
        df_h0["da"] = np.nan
        df_h0["db"] = np.nan
        df_h0["dE"] = np.nan
        df_h0["dC"] = np.nan
        
        df_h0.to_csv(f"csv_2/{name}_0H.csv", index=False)
        df_h240.to_csv(f"csv_2/{name}_240H.csv", index=False)
        df_h530.to_csv(f"csv_2/{name}_530H.csv", index=False)
        
    except Exception as e:
        raise e
```

```python
from os import listdir

for f in listdir("."):
    if ".xlsx" in f:
        try:
            split_into_csv(f)
        except Exception as e:
            print(f)
            raise e
```
